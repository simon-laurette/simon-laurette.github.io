Avant tout : importer le fichier Excel (format *\*.xls*) en cliquant sur la deuxième icone du menu en partant de la gauche (Dossier ouvert : "Open a notebook, load a file...").

*Remarque pratique :*
Ce Notebook détaille comment importer un fichier Excel via Basthon. Ainsi, il présuppose qu'on travaille avec les modules installés par défaut avec Basthon. Ici on travaillera avec les modules `pandas` et `xlrd` : c'est pourquoi on travaille avec des fichiers Excel *\*.xls* et non des fichiers *\*.xlsx*. Pour obtenir un fichier *\*.xls*, il suffit, sur Excel ou OpenOffice, d'enregistrer sous et de choisir le format "Classeur Excel 97-2003".
Pour ceux qui travaillent sour Spyder ou Jupyter Notebook, on peut aussi lire des fichiers *\*.xlsx*, mais vous devrez probablement installer également le module `openpyxl`.


```python
import pandas as pd
import xlrd

#ici le fichier importé est "determinationg.xls"
df = pd.read_excel("determinationg.xls") #df contient toutes les données du tableur
```

```python
#On peut visualiser le fichier 
df
```

Une fois que les données ont été importées, on peut les récupérer sous forme de tableaux.

```python
#On récupère les différentes colonnes sous forme de tableaux avec la syntaxe suivante :
deltat=df.values[:,0] #toute les lignes de la colonne 0
u_deltat=df.values[:,1]
h=df.values[:,2]
u_h=df.values[:,3]
```

```python
#Observons ce que contiennent ces tableaux
deltat
```

Et nous pouvons désormais effectuer un tracé :

```python
import matplotlib.pyplot as plt

plt.figure()
plt.errorbar(h,deltat,xerr=u_h,yerr=u_deltat,fmt='.',capsize=4) #fmt permet de définit le format de tracé (des points ici) et capsize le format des barres d'incertitude
plt.grid()
plt.xlabel("Hauteur de lâcher (m)")
plt.ylabel("Temps de chute (s)")
plt.show()
```
