> **Q14**.
>
> Mettre en œuvre l’algorithme de Monte-Carlo sur Python. On observera l’histogramme associé aux valeurs de la grandeur de sortie, et on calculera l’incertitude-type associée

```python
h=1 ### Remplissez avec vos valeurs
Delta_h=2e-2 ### (demi largeur de la plage d'incertitude)
Deltat=0.41 ### (valeur du temps de chute)
Delta_Deltat =0.1 ### (demi largeur de la plage d'incertitude associée)

g=2*h/Deltat**2

simu_h=np.random.uniform(h-Delta_h,h+Delta_h,10000)
simu_Deltat=np.random.uniform(Deltat-Delta_Deltat,Deltat+Delta_Deltat,10000)
simu_g=2*simu_h/simu_Deltat**2

plt.figure()
plt.hist(simu_g,bins=100)
plt.show()

print("g= ",g," m/s²")
print("u(g)= ",np.std(simu_g,ddof=1)," m/s²")
```
