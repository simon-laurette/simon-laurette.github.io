# Prise en compte du lycée d'origine


```python
## import données
import pandas as pd
import numpy as np

df=pd.read_excel("indicateurs.xls")
nom_lycee=df.values[:,0]
annee=df.values[:,1]
ville=df.values[:,2]
code=df.values[:,3]
reussite=df.values[:,4]
mentions=df.values[:,5]


#moyenne sur 3 ans

recap=[[nom_lycee[0],ville[0],0,0,0]] #colonne 3: reussite moyenne sur 3 ans, colonne 4:mentions moyennes ; colonne 5 : somme des précédentes

n=0.0
j=0
for i in range(len(nom_lycee)):
    if nom_lycee[i]==recap[j][0] and ville[i]==recap[j][1]:
        n+=1
        recap[j][2]+=reussite[i]
        recap[j][3]+=mentions[i]
    else :
        recap[j][2]=recap[j][2]/n
        recap[j][3]=recap[j][3]/n
        j+=1
        recap+=[[0,0,0,0,0]]
        n=1
        recap[j][0]=nom_lycee[i]
        recap[j][1]=ville[i]
        recap[j][2]+=reussite[i]
        recap[j][3]+=mentions[i]

recap[j][2]=recap[j][2]/n
recap[j][3]=recap[j][3]/n

for j in range(len(recap)):
    recap[j][4]=recap[j][2]+recap[j][3]

tableau=np.array(recap)
stockage_lycee=[tableau[i,0]+" "+tableau[i,1] for i in range(len(tableau))]
tableau2=tableau[np.array([float(tableau[i,4]) for i in range(len(tableau))]).argsort()] #classement par score croissant
tab3=np.array([tableau2[i,:] for i in range(len(tableau)-1,-1,-1)]) #décroissant

lycees=[tab3[i,0]+" "+tab3[i,1] for i in range(len(tab3))]
reus=[float(tab3[i,2]) for i in range(len(tab3))]
ment=np.array([float(tab3[i,3]) for i in range(len(tab3))])
sco=np.array([float(tab3[i,4]) for i in range(len(tab3))])

# utilitaire pratique

import tkinter as tk
from tkinter import ttk

def shiftmoyenne(score,moyenne_ini):
    """ fait la conversion de note en fonction du score du lycée"""
    if score>192:
        note=(20-11.0)/(20-5.6)*(moyenne_ini-20)+20
    elif score>177:
        note=(20-11.0)/(20-7.4)*(moyenne_ini-20)+20
    elif score>165:
        note=(20-11.0)/(20-9.2)*(moyenne_ini-20)+20
    elif score>145:
        note=moyenne_ini
    elif score>135:
        note=(20-11.0)/(20-12.8)*(moyenne_ini-20)+20
    elif score>120:
        note=(20-11.0)/(20-14.6)*(moyenne_ini-20)+20
    else :
        note=(20-11.0)/(20-16.4)*(moyenne_ini-20)+20 
    return note
        
def go():
    """ quant on appuie sur le bouton"""
    lyc=listeCombo.get()
    moyenne_ini=float(moyenne.get())
    for i in range(len(stockage_lycee)):
        if lyc==stockage_lycee[i]:
            jlycee=i
            break
    score=float(tableau[jlycee,4])
    newmean=str(shiftmoyenne(score,moyenne_ini))
    nouvellemoyenne.set(newmean)
    

root = tk.Tk() 
root.geometry('500x200')

labelChoix = tk.Label(root, text = "Choisir le lycée")
labelChoix.pack()


listeCombo = ttk.Combobox(root, values=stockage_lycee,width=300)
listeCombo.current(0)
listeCombo.pack()

labelMoy = tk.Label(root, text = "Moyenne Actuelle")
labelMoy.pack()

moyenne=tk.StringVar()
moyenne.set("")

moy=tk.Entry(root,textvariable=moyenne)
moy.pack()

bouton=tk.Button(root,text="Calculer la nouvelle moyenne",command=go)
bouton.pack(padx=5,pady=5)


nouvellemoyenne=tk.StringVar()
nouvellemoyenne.set("En Attente")

labelNew=tk.Label(root,textvariable=nouvellemoyenne)
labelNew.pack()

root.mainloop()
```
