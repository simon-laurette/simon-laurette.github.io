# Carte de champ d'un dipôle électrostatique

## Equipotentielles

L'équation polaire d'une équipotentielle est :

$$ r=\sqrt{\frac{p}{4\pi\varepsilon_0 V_0} \cos(\theta)}$$



```python
import matplotlib.pyplot as plt
import numpy as np
l_V0=np.linspace(-5,5,9)

ax = plt.subplot(111, polar=True)
ax.set_theta_zero_location("N")

for V0 in l_V0 :    
    if V0>0 :
        thet=np.linspace(-np.pi/2,np.pi/2,1000)
        r=np.sqrt(np.cos(thet)/V0)
        ax.plot(thet, r, label="V0="+str(V0)[0:4])

        
        
    if V0<0 :
        thet=np.linspace(np.pi/2+0.01,3*np.pi/2,1000)
        r=np.sqrt(np.cos(thet)/V0)
        ax.plot(thet,r,label="V0="+str(V0)[0:3]+"V")
   
    
plt.legend()
```

## Lignes de champ

L'équation d'une ligne de champ de $\vec{E}$ est :

$$r=K \sin^2(\theta)$$

avec $K>0$.



```python
ax = plt.subplot(111, polar=True)
ax.set_theta_zero_location("N")

l_K=np.linspace(0,1,6)
theta=np.linspace(-3.14/2,3*3.14/2,1000)
for K in l_K :
    ax.plot(theta,K*np.sin(theta)**2)
```

## Carte de champ complète

```python
import matplotlib.pyplot as plt
import numpy as np
l_V0=np.linspace(-5,5,9)

ax = plt.subplot(111, polar=True)
ax.set_theta_zero_location("N")

for V0 in l_V0 :    
    if V0>0 :
        thet=np.linspace(-np.pi/2,np.pi/2,1000)
        r=np.sqrt(np.cos(thet)/V0)
        ax.plot(thet, r,'--', label="V0="+str(V0)[0:4],color="black", linewidth=1)

        
        
    if V0<0 :
        thet=np.linspace(np.pi/2+0.01,3*np.pi/2,1000)
        r=np.sqrt(np.cos(thet)/V0)
        ax.plot(thet,r,'--',label="V0="+str(V0)[0:3]+"V",color="black", linewidth=1)
        
        
l_K=np.linspace(0,1,6)
theta=np.linspace(-3.14/2,3*3.14/2,1000)
for K in l_K :
    ax.plot(theta,K*np.sin(theta)**2,'-', color="black",linewidth=1)

ax.plot([0,3.14],[0.4,0.4],linewidth=4,color="blue")
ax.plot([0],[0.4],'^',markersize=10,color="blue")

plt.title("Carte de champ d'un dipôle électrostatique")
```
